```python
import pandas as pd
import os
```

Collect all the results in one dataframe 

```python
df = pd.DataFrame()

for result in os.listdir("./result"):
    result = 'result/' + result
    temp = pd.read_csv(result)
    df = pd.concat([df, temp])
```

# Best strategies

--> Talk about SQN

## Binance historical data (08/17/2017 - 12/25/2020)

```python
topSQN = df.sort_values('SQN', ascending=False).head(10)
topSQN['Rank'] = topSQN['SQN'].rank(method='min', ascending=False)
topSQN
```

Explain briefly what we see

```python
topSqnSqn = list(topSQN['SQN'])
topSqnProfit = list(round(topSQN['%'],3))
topSqnWin = list(topSQN['Total win'])
topSqnLoss = list(topSQN['Total loss'])

index = [i for i in range(10)] # COMPLETE WITH THE RANK !!
```

```python
topSqnBars = pd.DataFrame({'sqn': topSqnSqn, 'profit %': topSqnProfit}, index=index)

# topSqnBars.plot.bar();
topSqnBarsSplit = topSqnBars.plot.bar(rot=0, subplots=True) # PLOT X LEGEND
topSqnBarsSplit[1].legend(loc=2);
```

Talk about what we notice

What we will see next

```python
topSqnBars2 = pd.DataFrame({'win': topSqnWin, 'loss': topSqnLoss}, index=index)
topSqnBars2.plot.bar(rot=0); # PLOT X LEGEND
```
